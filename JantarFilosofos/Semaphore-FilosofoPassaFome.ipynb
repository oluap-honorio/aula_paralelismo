```python
from datetime import datetime 
from time import sleep, strftime, gmtime
from threading import Thread, Semaphore
from random import randrange
```

```python
class Filosofo:
    __n_forks = 0
    all_forks = 5
    sem = Semaphore(all_forks)
    
    def __init__(self, nome, num):
        self.nome = nome
        self.num = num
        self.cnt_pensou = 0
        self.cnt_comeu = 0
        self.cnt_garfo = 0
        
    def  eat(self):
        print("\n[{}][{}:{}] Iniciou a comer."
              .format(datetime.now().strftime('%d-%m-%Y %H:%M:%S'),self.nome,self.num))
        sleep(randrange(1,5))# indeterminismo
        print("\n[{}][{}:{}] Terminou de comer."
              .format(datetime.now().strftime('%d-%m-%Y %H:%M:%S'),self.nome,self.num))
        self.cnt_comeu += 1
        pass
    
    def  put_fork(self):
        if self.__n_forks > 0:
            self.__n_forks -= 1
            Filosofo.all_forks += 1
            self.sem.release()
            if self.__n_forks == 1:
                print("\n[{}][{}:{}] Soltou o primeiro garfo."
                      .format(datetime.now().strftime('%d-%m-%Y %H:%M:%S'),self.nome,self.num))
            elif self.__n_forks == 0:
                print("\n[{}][{}:{}] Soltou o segundo garfo."
                      .format(datetime.now().strftime('%d-%m-%Y %H:%M:%S'),self.nome,self.num))
        pass
    
    def  take_fork(self):
        while(Filosofo.all_forks < 1):
            if self.__n_forks == 0:
                print("\n[{}][{}:{}] Aguardando o primeiro garfo."
                  .format(datetime.now().strftime('%d-%m-%Y %H:%M:%S'),self.nome,self.num))
            elif self.__n_forks == 1:
                print("\n[{}][{}:{}] Aguardando o segundo garfo."
                  .format(datetime.now().strftime('%d-%m-%Y %H:%M:%S'),self.nome,self.num))
            sleep(randrange(1,5))# indeterminismo                
        self.sem.acquire()            
        self.cnt_garfo += 1
        self.__n_forks += 1
        Filosofo.all_forks -= 1
        if self.__n_forks == 1:
            print("\n[{}][{}:{}] Pegou o primeiro garfo."
                  .format(datetime.now().strftime('%d-%m-%Y %H:%M:%S'),self.nome,self.num))
        elif self.__n_forks == 2:
            print("\n[{}][{}:{}] Pegou o segundo garfo."
                  .format(datetime.now().strftime('%d-%m-%Y %H:%M:%S'),self.nome,self.num))
        pass
    
    def  think(self):
        print("\n[{}][{}:{}] Está pensando..."
              .format(datetime.now().strftime('%d-%m-%Y %H:%M:%S'),self.nome,self.num))
        self.cnt_pensou += 1
        sleep(randrange(1,5))# indeterminismo
        print("\n[{}][{}:{}] Ficou faminto."
              .format(datetime.now().strftime('%d-%m-%Y %H:%M:%S'),self.nome,self.num))
        pass

    def exist(self):
#       Ciclos de realizações
        for i in range(1,200):
            self.think();
            self.take_fork();
            self.take_fork();
            self.eat();
            self.put_fork();
            self.put_fork();
            print("\n")
            print(self.__dict__)
```

```python
f1 = Filosofo("FILOSOFO",1)
f2 = Filosofo("FILOSOFO",2)
f3 = Filosofo("FILOSOFO",3)
f4 = Filosofo("FILOSOFO",4)
f5 = Filosofo("FILOSOFO",5)

t1 = Thread(target=f1.exist)
t2 = Thread(target=f2.exist)
t3 = Thread(target=f3.exist)
t4 = Thread(target=f4.exist)
t5 = Thread(target=f5.exist)

t1.start()
t2.start()
t3.start()
t4.start()
t5.start()
```
